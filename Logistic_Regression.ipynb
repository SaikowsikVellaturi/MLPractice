```python
from sklearn.model_selection import train_test_split
```

```python
train_val_df,test_df=train_test_split(raw_df,test_size=0.2,random_state=42)
train_df,val_df=train_test_split(train_val_df,test_size=0.25,random_state=42)
```

```python
print(train_df.shape)
print(val_df.shape)
print(test_df.shape)
```

```python
year=pd.to_datetime(raw_df.Date).dt.year
train_df=raw_df[year<2015]
val_df=raw_df[year==2015]
test_df=raw_df[year>2015]
```

```python
print(train_df.shape)
print(val_df.shape)
print(test_df.shape)
```

```python
input_cols=list(train_df.columns)[1:-1]
target_col='RainTomorrow'
```

```python
train_inputs=train_df[input_cols]
train_targets=train_df[target_col]
val_inputs=val_df[input_cols]
val_targets=val_df[target_col]
```

```python
test_inputs=test_df[input_cols]
test_targets=test_df[target_col]
```

```python
import numpy as np
```

```python
numeric_cols=train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols=train_inputs.select_dtypes(include='object').columns.tolist()
```

```python
numeric_cols
```

```python
train_inputs[categorical_cols].nunique()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(strategy='mean')
```

```python
raw_df[numeric_cols].isna().sum()
```

```python
train_inputs[numeric_cols].isna().sum()
```

```python
imputer.fit(raw_df[numeric_cols])
```

```python
train_inputs[numeric_cols]=imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=imputer.transform(val_inputs[numeric_cols])


test_inputs[numeric_cols]=imputer.transform(test_inputs[numeric_cols])
```

```python
raw_df[numeric_cols].describe()
```

```python

from sklearn.preprocessing import MinMaxScaler

scaler_ = MinMaxScaler()

train_inputs[numeric_cols] = scaler_.fit_transform(train_inputs[numeric_cols])

# Transform the validation and test data using the same scaler
val_inputs[numeric_cols] = scaler_.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler_.transform(test_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols]
```

```python

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
```

```python
raw_df2=raw_df[categorical_cols].fillna('Unknown')
```

```python
encoder.fit(raw_df2)
```

```python
encoder.categories_
```

```python
encoded_cols=list(encoder.get_feature_names_out(categorical_cols))
```

```python
encoded_cols
```

```python



train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols].fillna('Unknown'))
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols].fillna('Unknown'))
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols].fillna('Unknown'))


```

```python
test_inputs
```

```python
train_inputs.to_parquet('train_inputs.parquet')
val_inputs.to_parquet('val_inputs.parquet')
test_inputs.to_parquet('test_inputs.parquet')
```

```python
pd.DataFrame(train_targets).to_parquet('train_targets.parquet')
pd.DataFrame(val_targets).to_parquet('val_targets.parquet')
pd.DataFrame(test_targets).to_parquet('test_targets.parquet')
```
